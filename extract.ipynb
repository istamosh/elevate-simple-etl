```python
import requests
from bs4 import BeautifulSoup as bs
import time
import json
import os
```

```python
# list url pagination yang akan di-scrape

def build_urls():
    return ['https://fashion-studio.dicoding.dev/'] + [f'https://fashion-studio.dicoding.dev/page{i}' for i in range(2, 51)]
urls = build_urls()
print(urls)
```

```python
FILE = "extracted_data.json"

# cek apakah file sudah ada, jika ada load data dari file, jika tidak scrape data dari web hanya sekali dan simpan ke file
if os.path.exists(FILE):
    with open(FILE, "r", encoding="utf-8") as f:
        data = json.load(f)
        print(f'File {FILE} ditemukan, data berhasil dimuat dari file.')
else:
    # siapkan list untuk menampung data
    data = []

    for url in urls:
        r = requests.get(url, timeout=10)   # request ke setiap url
        r.raise_for_status()  # pastikan request berhasil

        # parsing response dengan BeautifulSoup
        soup = bs(r.text, 'html.parser')

        # jadikan card dengan class product-details sebagai items list
        items = soup.find_all(class_="product-details")

        # ekstrak data dari setiap item
        for item in items:
            title = item.find(class_="product-title").get_text(strip=True)
            price = item.find(class_="price").get_text(strip=True)

            ps = item.find_all('p')
            rating = ps[0].get_text(strip=True)
            color = ps[1].get_text(strip=True)
            size = ps[2].get_text(strip=True)
            gender = ps[3].get_text(strip=True)

            # simpan data ke list
            data.append({
                'title': title,
                'price': price,
                'rating': rating,
                'color': color,
                'size': size,
                'gender': gender
            })

        print(f'Sukses scrape {url}')
        time.sleep(1)  # delay 1 detik untuk menghindari rate limit

    # simpan data ke file JSON
    with open("extracted_data.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)
```

```python
# tampilkan 3 data pertama
print(json.dumps(data[:3], indent=2))
```
