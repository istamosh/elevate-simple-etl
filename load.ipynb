```python
import csv
import json
import pandas as pd
```

```python
with open("transformed_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

with open("final_data.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)
```
